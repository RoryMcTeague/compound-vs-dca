```python
''' Compound Interest vs. Dollar-Cost Averaging (DCA)

In this project, we compare two investment strategies:

1. **Compound Interest**: A one-time lump sum investment that grows over time.
2. **Dollar-Cost Averaging (DCA)**: Regularly investing a fixed amount over time.

We'll simulate both strategies and visualize how they perform over time.'''
```

```python
# Importing necessary libraries for plotting and handling date/time functionality

import matplotlib.pyplot as plt  # For creating visualizations and plots
import datetime  # For working with date and time objects
```

```python
#Defining Variables

#Lumpsum for Compound Interest
initial_investment = float(input("Enter your initial investment amount: "))

#Monthly Contribution for DCA
monthly_investment = float(input("Enter your monthly contribution: "))

#Annual return converted to decimal
annual_return = float(input("Enter expected annual return (in percentage, e.g. 7 for 7%): ")) / 100

#Time period
years = int(input("Enter the number of years for investment: "))
```

```python
#Defining Functions
# Compound Interest
def compound_interest(principal, rate, years):
    return principal * (1 + rate) ** years

# Dollar Cost Averaging
def dollar_cost_averaging(monthly, rate, years):
    total = 0
    for month in range(1, 12 * years + 1):
        total += monthly * (1 + rate / 12) ** ((years * 12 - month))
    return total
    
```

```python
#Simulating Over Time Period
#initialise lists for simulation
compound_values = []
DCA_values = []
time_points = list(range(1, years + 1))

#run simulations and store results
for year in time_points:
    compound_values.append(compound_interest(initial_investment, annual_return, year))
    DCA_values.append(dollar_cost_averaging(monthly_investment, annual_return, year))
    
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))

# Plot Compound Interest
plt.plot(time_points, compound_values, label="Compound Interest", color="Blue")

# Plot Dollar Cost Averaging
plt.plot(time_points, DCA_values, label="Dollar Cost Averaging", color="Red")

# Labeling the axes
plt.xlabel("Years")
plt.ylabel("Portfolio Value ($)")

# Title of the plot
plt.title("Compound Interest vs Dollar Cost Averaging")

# Displaying the legend
plt.legend()

# Enabling the grid
plt.grid(True)

# Display the plot
plt.show()
```
